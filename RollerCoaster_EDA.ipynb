```python
df = df[['coaster_name', 
         #'Length', 'Speed', 
         'Location', 'Status', 
        # 'Opening date','Type',
         'Manufacturer', 
         #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
         #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
         #'height_value', 'height_unit',
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df['opening_date_clean']
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df[df.duplicated(subset=['coaster_name'], keep=False)].sort_values(by='coaster_name')
```

```python
#'~' it inverse the value

df = df.loc[~df.duplicated(subset=['coaster_name', 'location', 'opening_date_clean'], keep=False)].reset_index(drop=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.drop('height_ft',axis=1, inplace=True)
```

```python
df.drop('gforce_clean',axis=1, inplace=True)
```

```python
df.fillna({'manufacturer': 'Unknown', 'status' : 'Unknown'}, inplace=True)
```

```python
df.shape
```

```python
df['speed_mph'] = df['speed_mph'].fillna(df['speed_mph'].median())
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['latitude','longitude'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['opening_date_clean'] = df['opening_date_clean'].fillna(df['year_introduced'])
```

```python
ax = (
    df.query('location != "Other"')
      .groupby('location')['speed_mph']
      .agg(['mean','count'])
      .query('count > 10')
      .sort_values('mean')['mean']
      .plot(kind='barh', figsize=(12,6))
)

# Labels & title
ax.set_xlabel('Average Coaster Speed (mph)')
ax.set_ylabel('Location',)
ax.set_title('Average Coaster Speed by Location', fontsize=16)


# Improve spacing
plt.tight_layout()

plt.show()
```

```python

# Prepare data
data = (
    df.query('location != "Other"')
      .groupby('location')['speed_mph']
      .agg(['mean','count'])
      .query('count > 10')
      .sort_values('mean')
      .reset_index()
)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(
    data=data,
    x='mean',
    y='location',
    hue='location',
    palette='coolwarm'
)

# Labels
plt.xlabel('Average Coaster Speed (mph)', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.title('Average Coaster Speed by Location', fontsize=14, pad=15)

plt.tight_layout()
plt.show()
```
